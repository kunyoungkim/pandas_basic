```python
import pandas as pd

data = {
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['New York', 'Los Angeles', 'Chicago']
}

df = pd.DataFrame(data)

# 특정 행과 열에 접근
print(df.loc[1, 'Age'])  # 출력: 30

# 특정 행 전체를 선택
print(df.loc[1])
# 출력:
# Name       Bob
# Age         30
# City    Los Angeles
# Name: 1, dtype: object

# 특정 열 전체를 선택
print(df.loc[:, 'City'])
# 출력:
# 0       New York
# 1    Los Angeles
# 2       Chicago
# Name: City, dtype: object

# 여러 행과 열 선택
print(df.loc[0:1, ['Name', 'City']])
# 출력:
#    Name         City
# 0  Alice     New York
# 1    Bob  Los Angeles
```

```python
import pandas as pd

# 예제 데이터프레임 생성
data = {
    'A': [1, 2, 3],
    'B': [4, 5, 6]
}
df = pd.DataFrame(data, index=['a', 'b', 'c'])

df
# print("Original DataFrame:")
# print(df)

# # 인덱스 리셋
# df_reset = df.reset_index()

# print("\nDataFrame after reset_index:")
# df_reset
```

```python
# 인덱스 리셋하면서 새로운 인덱스 열의 이름을 지정
df.reset_index(names='NewIndexName')
```

```python
df_reset = df.reset_index()
df_reset
```

```python
df_reset = df.reset_index(drop=True)
df_reset
```

```python
# 다중 인덱스 데이터프레임 생성
arrays = [['bar', 'bar', 'baz', 'baz'], ['one', 'two', 'one', 'two']]
index = pd.MultiIndex.from_arrays(arrays, names=['first', 'second'])
df_multi = pd.DataFrame({'A': [1, 2, 3, 4], 'B': [5, 6, 7, 8]}, index=index)

df_multi
```

```python
index
```

```python
# 특정 레벨만 리셋
df_multi_reset = df_multi.reset_index(level='first')

print("\nMultiIndex DataFrame after resetting 'first' level:")
df_multi_reset
```
